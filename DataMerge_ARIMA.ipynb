```python
from sklearn.model_selection import train_test_split

#SPLIT THE DATA INTO 70 AND 30 FOR forecasting
#trainend = round(df.shape[0]*0.7)
#trainend2 = round(df_2.shape[0]*0.7)

#train = df[0:trainend]
#test = df[trainend:df.shape[0]]

#we test for a single dataset first
#train2 =  df_2[0:trainend2]
#test2 = df[trainend2:df_2.shape[0]]


train, test = train_test_split(df, test_size=0.3)

train2, test2 = train_test_split(df_2, test_size=0.3)
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
#have to do the train_test split beforehand all the data that is handed to the function will be trained on
#
#train2['date']
y = train2["prec"]
exog = train2.drop(["prec", "date","Identifier"], axis = 1)
model = ARIMA(endog = y, exog = exog,order=(1,1,1))
#model = ARIMA(endog = y, exog = exog,dates=train2["d"], order=(1,1,2))
model_fit = model.fit()
```

```python
exog_new = test2.drop( ["prec", "date", "Identifier"], axis = 1)

pred = model_fit.forecast(len(test2), exog = exog_new)
pred = pred.to_frame()


# If it predicts negative precipitation make it 0
for i in range(0,pred.shape[0]):
    if(pred.iat[i,0] < 0):
        pred.iat[i,0]=0
#pred = pred.reset_index().drop(["index"], axis = 1)

# Compare to the results
test2= test2.sort_index()

plt.plot(test2["prec"])
plt.plot(pred)
plt.xlabel("Timestamp")
plt.ylabel("prec")

plt.legend(["train", "predicted"])
plt.show()
mean_squared_error(test2["prec"], pred)
```
